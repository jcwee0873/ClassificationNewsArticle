## 빅카인즈 데이터를 바탕으로 기사 크롤링
- 빅카인즈에서 제공하는 기사 제목, URL을 바탕으로 기사 본문 크롤링

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_excel('./news_data/서울신문.xlsx')
```

```python
columns = ['일자', '제목', 'URL']
news_list = df[columns]
```

```python
news_list.URL[22]
```

```python
news_list['본문'] = np.nan
```

```python
news_list
```

```python
def get_news_body(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content.decode('euc-kr'), 'html.parser')
    if 'http://www' in url : 
        text = soup.find('div', attrs = {'class' : 'S20_v_article'}).text

    elif 'http://nownews' in url :
        text = soup.find('div', attrs = {'class':'now_txt1'}).text

    elif 'http://go' in url :
        text = soup.find('div', attrs = {'class':'atic_txt1'}).text
    print(text.replace('\n', '')[:10])
    return text
```

```python
for i, url in enumerate(news_list['URL']):
    print(i,'번째')
    try :
        news_list['본문'][i] = get_news_body(url)
    except :
        pass
```

```python
news_list.to_csv('./seoul.csv')
```
